```python
import re
from konlpy.tag import Okt
#정규식, 한국어 전처리할 때 쓰는 형태소 분석기
```

```python
word_list=[]#단어 리스트라는 빈 리스트 만들기
word_list=df1['text']#text의 내용을 word_list에 넣기
word_list=word_list.reset_index(drop=True)
word_list
```

```python
#본문 내용이 들어가면 정규화하는 함수 만들기기
def pop_word(text):
    korean = re.compile('[^가-힣]') 
    result = korean.sub(' ', text) 
    return result
for i in range(0,len(word_list)):#기사가 많으면 반복하면서 차례대로 쌓일 텐데 내용이 1개라서 큰 의미가 없다. 나중에 크롤링을 보완하면 더 좋을 듯듯
    word_list[i] = pop_word(word_list[i])
```

```python
word_list#확인
```

```python
Text=word_list
#데이터 프레임 형태로 보여주기
df2=pd.DataFrame([Text])
df2= df2.transpose()#전치치
```

```python
df2
```

```python
from tqdm import tqdm#상태 보여주는 것
okt = Okt()#트위터 형태소 분석기
```

```python
nouns = [] 
for x in tqdm(Text):
    try:
        nouns.append(okt.nouns(x))
    except:
        nouns.append(["None"])
```

```python
# 명사 추출 결과값 열에 추가
df2_no= df2.copy()
df2_no["Noun"] = nouns
```

```python
nouns
```

```python
def change(text):
    text=','.join(text)
    text_list=text.split(',')
    result=[]
    for b in text_list:
        result.append(b)
    result=','.join(result)
    return result
```

```python
df2_no["Noun"]=df2_no["Noun"].map(lambda x: change(x))
```

```python
def remove(text):
    text_list=text.split(',')
    result=[]
    for n in text_list:
        if len(n)>1:
            result.append(n)
    result=' '.join(result)
    return result

df2_no["Noun"]=df2_no["Noun"].map(lambda x: remove(x))
```

```python
noun=df2_no["Noun"]
```

```python
import nltk

for i, document in enumerate(noun):
  clean_words = [] 
  for word in nltk.tokenize.word_tokenize(document):
    if word not in np.array(stopword): #불용어 제거
     clean_words.append(word)
  noun[i] = ' '.join(clean_words)
```

```python
#다시 데이터 프레임으로
df3 = pd.DataFrame(noun)
df3.rename(columns={'Noun':'Key'},inplace=True)
df3.to_csv('부족한전처리.csv',encoding='utf-8',index=False)
```

```python
#빈도수 함수 만들기
freq = df3['Key']#df3의 내용에서 확인 하기 위해 따로 변수 지정

#문자열 끊기,리스트 형식으로 저장
wds = [str(i).split(" ")for i in freq]

#함수 만들기
def freq_word(wds):
  fr_w = dict()
  for wd in wds:
    if fr_w.get(wd):
      fr_w[wd]+=1#같은 거 찾으면 +1
    else :
      fr_w[wd] = 1 #그대로 진행
  fre_df = pd.DataFrame.from_dict(fr_w,orient='index')#딕셔너리를 데이터 프레임으로 변환
  fre_df.columns = ['frequency']
  fre_df = fre_df.sort_values(['frequency'],ascending=False)#오름차순으로 정렬 
  return fr_w, fre_df   
 
```
